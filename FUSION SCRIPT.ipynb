```python
import pandas as pd
```

```python
data = pd.read_csv("InmobiliariasA-I.csv") 
```

```python
data.head()
```

```python
data['Inmobiliaria'].value_counts()
```

```python
data.loc[lambda data: data['Inmobiliaria'].str[0] == 'J']['Inmobiliaria'].value_counts()
```

```python
data.loc[lambda data: data['Inmobiliaria'].str[0] != 'J']['Inmobiliaria'].value_counts()
```

```python
dataA_I =  data.loc[lambda data: data['Inmobiliaria'].str[0] != 'J']
dataA_I.head()
```

```python
data2 = pd.read_csv("InmobiliariasJ-L.csv") 
data2.head()
```

```python
data_temp = data2.loc[lambda data2: data2['Inmobiliaria'].str[0] != 'M']
data_temp = data_temp.loc[lambda data_temp: data_temp['Inmobiliaria'].str[0] > 'I']
dataJ_L= data_temp.loc[lambda data_temp: data_temp['Inmobiliaria'].str[0] != 'e']
```

```python
dataJ_L['Inmobiliaria'].value_counts()
```

```python
data3 = pd.read_csv("InmobiliariasM-Q.csv") 
data3.head()
```

```python
data_temp = data3.loc[lambda data3: data3['Inmobiliaria'].str[0] != 'R']
data_temp['Inmobiliaria'].value_counts()
```

```python
dataM_Q = data_temp
```

```python
data4 = pd.read_csv("InmobiliariasR-U.csv") 
data4.head()
```

```python
data_temp = data4.loc[lambda data4: data4['Inmobiliaria'].str[0] != 'V']
data_temp['Inmobiliaria'].value_counts()
```

```python
dataR_U = data_temp
```

```python
data5 = pd.read_csv("InmobiliariasV-Z.csv") 
data5.head()
```

```python
data5['Inmobiliaria'].value_counts()
```

```python
data6 = pd.read_csv("InmobiliariasZ.csv") 
data6.head()
```

```python
data6['Inmobiliaria'].value_counts()
```

```python
data_inmo_final = pd.concat([dataA_I, dataJ_L, dataM_Q, dataR_U, data5, data6])
data_inmo_final['Inmobiliaria'].value_counts()
```

```python
data_inmo_final.to_csv('Data_Inmobiliarias.csv',index=False)
```
