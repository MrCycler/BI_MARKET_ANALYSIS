```python
#importación d elibrerías
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.chrome.service import Service

from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait

import base64
import wget
import requests
import unicodecsv as csv
import datetime
import re

#total_depas
total_depas =[]
```

```python
#save a string array in a csv file
def save_csv(depas):
    with open('Inmobiliarias_rawdata.csv', 'wb') as f:
            dict_writer = csv.DictWriter(
                f,
                [
                'Inmobiliaria', 
                'Presentation',
                'Imagen_Inmobiliaria',
                'URL_Inmobiliaria',
                'Proyecto', 
                'Distrito_Proyecto',
                'URL_Proyecto',
                'Direccion_Proyecto',
                'Tipo_Proyecto',
                'Estado_Proyecto',
                'Release_Date_Proyecto',
                'Banco_Proyecto',
                'Modelo_Vivienda', 
                'NroPiso_Vivienda', 
                'NroDormitorios_Vivienda', 
                'Area_Vivienda', 
                'Precio_Vivienda', 
                'Unidades_Vivienda', 
                'URL_Image_Vivienda', 
                'Areas_Comunes_Vivienda',
                ],
                encoding='utf-8')
            dict_writer.writeheader()
            dict_writer.writerows(depas)
            
def get_title(article):
    return article.find('h3').text

def get_main_web(article):
    return article.find('a', {'class' : 'Realstate-link'})['href']

def get_presentation(article):
    return article.find('p').text

def get_presentation(article):
    return article.find('p').text

def get_url_image(article):
    figure = article.find('figure', {'class' : 'Realstate-img'})
    return figure.find('img')['src']

def text_filter(text):
    filter_text = text.replace(' ','')
    filter_text = filter_text.replace('\n','') 
    filter_text = filter_text.replace('\r','')
    return filter_text

def get_project_info(url, title_inmo, presentation_inmo, url_image_inmo, project_name, project_distrito, inmo_url):
    service = Service('C:/Users/pc/Desktop/SCRAP-INMO/chromedriver')
    service.start()
    driver = webdriver.Remote(service.service_url)
    driver.get(url)
    driver.implicitly_wait(50)
    driver.find_elements_by_class_name("ct-ficha-proyecto")
    soup_projects = BeautifulSoup(driver.page_source,'lxml')
    project_addresses = soup_projects.findAll('p', {'class' : 'Project-header-address'})
    project_address = ""
    for address in project_addresses:
        project_address = project_address + address.text
    print(project_address) #project address(C)
    project_box = soup_projects.find('div', {'class' : 'bx-data-project'})
    project_tds = project_box.findAll('td')
    project_type = text_filter(project_tds[1].text)
    print(project_type)#project type(C)
    project_status = text_filter(project_tds[7].text)
    print(project_status)#project status(C)
    project_rdate = text_filter(project_tds[9].text)
    print(project_rdate)#project release date(C)
    project_bank = text_filter(project_tds[11].text)
    print(project_bank)#project bank(C)
    areas_comunes =''
    areas_soup = soup_projects.find('ul', {'class' : 'Project-areas-list'})
    if (areas_soup is None):
        areas_comunes =''
    else:
        areas_scrap = areas_soup.findAll('li')
        for comun in areas_scrap:
            areas_comunes = areas_comunes + str(comun.text)+','
    print(areas_comunes)#areas comunes (C)
    project_models = soup_projects.findAll('div', {'class' : 'Project-available-model'})
    #obtenemos departamentos
    for model_data in project_models:
        print('-------------------Departamento----------------')
        model = text_filter(model_data.find('span', {'class' : 'name_tipology'}).text)
        print(model)#model(C)
        if(model_data.find('span', {'class' : 'num_pisos'})is None):
            nro_piso = ''
        else:
            nro_piso = text_filter(model_data.find('span', {'class' : 'num_pisos'}).text)
        print(nro_piso)#nropiso(C)
        if(model_data.find('span', {'class' : 'bedroom'})is None):
            nro_dormitorios = ''
        else:
            nro_dormitorios = text_filter(model_data.find('span', {'class' : 'bedroom'}).text)
            print(nro_dormitorios)#nro_dormitorios(C)
        area = text_filter(model_data.find('span', {'class' : 'area'}).text)
        print(area)#area(C)
        price = text_filter(model_data.find('span', {'class' : 'price'}).text)
        print(price)#price(C)
        unidades = text_filter(model_data.find('span', {'class' : 'text-unit'}).text)
        temp_unidades = re.findall(r'\d+', unidades)
        print(int(temp_unidades[0]))#unidades(C)
        preview = model_data.find('div', {'class' : 'Project-available-preview'})
        if(preview.find('img') is None):
            img_url = ''
        else:
            img_url = preview.find('img')['src']
        print(img_url)#apartment_model_url(C)
        print('-------------------------------------------')
        total_depas.extend([
            {
                'Inmobiliaria':str(title_inmo), 
                'Presentation':str(presentation_inmo),
                'Imagen_Inmobiliaria': str(url_image_inmo),
                'URL_Inmobiliaria':str(inmo_url),
                'Proyecto':str(project_name), 
                'Distrito_Proyecto':str(project_distrito),
                'URL_Proyecto':str(url),
                'Direccion_Proyecto':str(project_address),
                'Tipo_Proyecto':str(project_type),
                'Estado_Proyecto':str(project_status),
                'Release_Date_Proyecto':str(project_rdate),
                'Banco_Proyecto':str(project_bank),
                'Modelo_Vivienda':str(model), 
                'NroPiso_Vivienda':str(nro_piso), 
                'NroDormitorios_Vivienda':str(nro_dormitorios), 
                'Area_Vivienda':str(area), 
                'Precio_Vivienda':str(price), 
                'Unidades_Vivienda':int(temp_unidades[0]), 
                'URL_Image_Vivienda':str(img_url), 
                'Areas_Comunes_Vivienda':str(areas_comunes), 
            }
        ])
    save_csv(total_depas)
        
def get_projects(url, title, presentation, url_image):
    service = Service('C:/Users/pc/Desktop/SCRAP-INMO/chromedriver')
    service.start()
    driver = webdriver.Remote(service.service_url)
    driver.get(url)
    driver.implicitly_wait(50)
    driver.find_elements_by_class_name("RealestateProjects")
    soup_projects = BeautifulSoup(driver.page_source,'lxml')
    projects = soup_projects.findAll('div', {'class' : 'SearchResult-project-data'})
    for project in projects:
        print('----------------------Proyecto------------------------')
        project_url = project.find('a')['href']
        print(project_url) #url(C)
        project_name = project.find('a').text
        print(project_name) #name(C)
        project_distrito = project.findAll('span')[0].text
        print(project_distrito) #distrito(C)
        print(project.find('span', {'class' : 'price'}).text) #min_price
        get_project_info(project_url, title, presentation, url_image, project_name, project_distrito, url)
        print('----------------------------------------------------------')


def iterate_inmobiliarias(url): 
    for addin in range(ord('A'),ord('Z')+1):    
        url_inmolet = url +'/'+chr(addin)
        print('----------------------URL Scrapeada-----------------------')
        print(url_inmolet)
        print('----------------------------------------------------------')
        content_page = requests.get(url_inmolet).content
        soup_page = BeautifulSoup(content_page,'lxml')
        articles_inmo = soup_page.findAll('article', {'class' : 'Realstate'})
        for article in articles_inmo:
            title = get_title(article)
            main_web = get_main_web(article)
            presentation = get_presentation(article)
            url_image = get_url_image(article)
            print('----------------------Inmobiliaria------------------------')
            print(title)#inmobiliaria(C)
            print(main_web)#web(C)
            print(presentation)#presentation(C)
            print(url_image)#url_image(C)
            get_projects(main_web, title, presentation, url_image)
  
```
