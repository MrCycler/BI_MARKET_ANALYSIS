```python
lista_inmobiliarias = data['Inmobiliaria'].unique()
len(lista_inmobiliarias)
```

```python
proyectos_array = []
inmobiliarias_array = []
#Listado de inmobiliarias y proyectos
for inmobiliaria in lista_inmobiliarias:
  listado_proyectos = data.loc[data['Inmobiliaria']==inmobiliaria]['Proyecto'].unique()
  area_inmo = 0
  valor_inmo = 0
  numero_proyectos = 0
  for proyecto in listado_proyectos:
    listado_depas = data.loc[data['Proyecto']==proyecto]
    precios = listado_depas['Precio_Vivienda'].values
    areas = listado_depas['Area_Vivienda'].values
    unidades = listado_depas['Unidades_Vivienda'].values
    i=0;
    area_total = 0;
    valor_total = 0;
    for precio in precios:
      area = float(areas[i].replace('m2',''))
      if'$'in precio:
        precio_temp = precio.replace('$','')
        precio_temp = precio_temp.replace(',','')
        precio_int = int(precio_temp)*3.38
      else:
        precio_temp = precio.replace('S','')
        precio_temp = precio_temp.replace('/','')
        precio_temp = precio_temp.replace(',','')
        precio_int = int(precio_temp)
      valor_total = valor_total + precio_int
      area_total = area_total + area*(unidades[i])
      i=i+1
    area_inmo = area_inmo + area_total
    valor_inmo = valor_inmo + valor_total 
    numero_proyectos = numero_proyectos + 1
    proyectos_array.append([proyecto, inmobiliaria, area_total, valor_total])
  inmobiliarias_array.append([inmobiliaria, numero_proyectos, area_inmo, valor_inmo])
```

```python
df_proyectos = pd.DataFrame(proyectos_array, columns = ['Proyecto','Inmobiliaria','Area Total', 'Valor Total (PEN)']) 
df_proyectos.head()
```

```python
df_inmobiliarias = pd.DataFrame(inmobiliarias_array, columns = ['Inmobiliaria','Número Proyectos','Area Total', 'Valor Total (PEN)']) 
df_inmobiliarias.head()
```
